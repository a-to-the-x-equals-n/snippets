`L1 Manhattan`

```python
import numpy as np

# Define the two vectors
v1 = np.array([5.1,7.3,-6.2,-1.3,9.4,-3.5])
v2 = np.array([7.2,3.1,4.1,-8.8,-5.3,1.5])

# Calculate the L1 norm (Manhattan distance)
distance = np.linalg.norm(v1 - v2, ord=1)

print(distance)
```

`L2 Euclidean`

```python
import numpy as np

v1 = np.array([5.1, 7.3, -6.2, -1.3, 9.4, -3.5])
v2 = np.array([7.2, 3.1, 4.1, -8.8, -5.3, 1.5])
distance = np.linalg.norm(x = v1 - v2, ord = 2)
distance
```

`Cosine Similarity`

```python
import numpy as np

# Define the two vectors
v1 = np.array([5.1,7.3,-6.2,-1.3,9.4,-3.5])
v2 = np.array([7.2,3.1,4.1,-8.8,-5.3,1.5])

# Calculate the magnitudes (L2 norm) of the vectors
magnitude_v1 = np.linalg.norm(x = v1, ord = 2)
magnitude_v2 = np.linalg.norm(x = v2, ord = 2)

# Calculate cosine similarity
cosine_similarity = v1 @ v2 / (magnitude_v1 * magnitude_v2)

cosine_similarity
```
