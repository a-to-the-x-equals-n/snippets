```python
def inliers(p, n):
    result = p ** n
    print(f"Inliers Probability: {result}")
    return result

def no_inliers(p, n):
    result = 1 - inliers(p, n)
    print(f"No Inliers Probability: {result}")
    return result

def iter_no_inliers(p, n, iter):
    no_match = no_inliers(p, n)  # This should use the result directly
    result = no_match ** iter
    print(f"Probability of No Inliers in {iter} Iterations: {result}")
    return result

def one_inlier_match(p, n, iter):
    result = 1 - iter_no_inliers(p, n, iter)
    print(f"Probability of At Least One Inlier Match in {iter} Iterations: {result}")
    return result

# Example usage:
p = 0.5  # 50% chance of a point being an inlier
n = 4     # Using the four-point algorithm
iterations = 55

one_inlier_match(p, n, iterations)
```
