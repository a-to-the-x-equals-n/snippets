```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

`cv2 is BGR`
---

```python
bird = cv2.imread("test.png")
plt.imshow(bird)
plt.title('BGR')
plt.show()
```

`slice np arrays`
---

```python
bird2 = bird[..., ::-1]
plt.imshow(bird2)
plt.title('RGB w/ Slicing')
plt.show()
```

`with cv2`
---

```python
bird2 = cv2.cvtColor(bird, cv2.COLOR_BGR2RGB)
plt.imshow(bird2)
plt.title('RGB w/ cv2')
plt.show()
```

`cv2 grayscale`
---

```python
bird_gray = cv2.cvtColor(bird, cv2.COLOR_BGR2GRAY)
plt.imshow(bird_gray, cmap = 'gray')
plt.show()
```
