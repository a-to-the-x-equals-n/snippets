```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

reference = image = cv2.imread('Bird0.png')
plt.imshow(X = image)
plt.show()
```

```python
sift = cv2.SIFT_create()
keypoints = sift.detect(reference, None)
image = cv2.drawKeypoints(reference, keypoints, image, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(X = image[..., ::-1])
plt.show()
```

```python
keypoints, descriptors = sift.detectAndCompute(reference, None)
print(descriptors.shape)
```
