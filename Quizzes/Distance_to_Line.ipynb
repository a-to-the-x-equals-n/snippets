`Distance to Point`
---

```python
import numpy as np


def point_distance(P, A, B):
    # P : the new point where the distance is being calculated
    # A : point of vector
    # B : second point of vector
    v = P - A
    n = B - A
    n_hat = n / np.sqrt(n @ n)
    C = (v @ n_hat) * n_hat + A
    return np.linalg.norm( P - C )
```

```python
P = np.array([-0.6,
              -6.9], dtype = np.float32)
A = np.array([3.0,
              6.3], dtype = np.float32)
B = np.array([0.2,
              1.6], dtype = np.float32)

print(point_distance(P, A, B))
```

`Vector Length`
---

```python
# "Length" = Magnitude
def vector_length(n):
    return np.sqrt(n @ n)
```

```python
vector = np.array([5.4, 0.1])
print(vector_length(vector))
```

`Vector X and Y`
---

```python
def vector_unit_x(v, axis_y = False):
    return vector[int(axis_y)] / np.linalg.norm(v)
```

```python
vector = np.array([-5.6, 7.2])
print(vector_unit_x(vector))
```

```python
vector = np.array([4.5, 7.1])
print(vector_unit_x(vector, axis_y = True))
```
