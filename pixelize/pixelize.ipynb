`Pixelize`
---

`code`

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pixelize(image: np.ndarray, bit: int = 32) -> np.ndarray:

    image_y, image_x, _ = image.shape
    kernel_y, kernel_x = image_y // bit, image_x // bit
    pixelized_image = np.zeros_like(a = image, dtype = np.uint8)

    for y in range(0, image_y, kernel_y):
        for x in range(0, image_x, kernel_x):
            # kernel
            pixel_window = image[y : y + kernel_y, x : x + kernel_x] 
            pixel_color = pixel_window.mean(axis = (0, 1), dtype = int)
            # fill pixel with averaged color
            pixelized_image[y : y + kernel_y, x : x + kernel_x] = pixel_color
    
    return np.clip(pixelized_image, 0, 255).astype(dtype = np.int32)
```

`choose image`

```python
# beach.jpg | boat.jpg | raccoon.jpg
raccoon = plt.imread('raccoon.jpg')
racocon = np.array(raccoon, dtype = np.int32)
```

`demo`

```python
plt.imshow(raccoon)
plt.title("original")
plt.show()

plt.imshow(pixelize(raccoon, bit = 64))
plt.title("64 bit")
plt.show()

plt.imshow(pixelize(raccoon, bit = 32))
plt.title("32 bit")
plt.show()

plt.imshow(pixelize(raccoon, bit = 16))
plt.title("16 bit")
plt.show()
```
